1. `Klassik dasturlash` orqali qilingan natija
---------

````python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt

# data
data = make_regression(n_samples=40, n_features=1, noise=10, random_state=42)
x_data, y_data = data[0], data[1]

def classic_program(x): return x*50

y_pred = [classic_program(x) for x in x_data]

# plotting
plt.scatter(x_data, y_data, label='Haqiqiy data sets')
plt.plot(x_data, y_pred, 'r', label='Dastur')
plt.xlabel('x')
plt.ylabel('Y')
plt.legend()
plt.title('x*50')
````

2. Vazn qo'llanilgan dastur

Bu kod bilan `w` qiymatini funksiyamizga qo'shib nuqta-nuqta bo'lib o'sib borgan grafikka mos bo'lgan to'g'ri chiziqni o'zgartirishni o'rgandik.

Lekin eng optimal qiymaatni bu yo'l bilan topa olmaymiz, chunki bu yerda `w` ga beradigan qiymatlarimiz juda ko'p. Shuning uchun bizga ```Machine Learning```yordamga keladi.

````python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt

# data
data = make_regression(n_samples=40, n_features=1, noise=10, random_state=42)
x_data, y_data = data[0], data[1]

w = 0.7

def vaznli_dastur(x, w): return x*w*50

y_pred = [vaznli_dastur(x, w) for x in x_data]

# plotting
plt.scatter(x_data, y_data, label='Haqiqiy data sets')
plt.plot(x_data, y_pred, 'r', label='Vaznli dastur')
plt.xlabel('x')
plt.ylabel('Y')
plt.legend()
plt.title(f'{w}x*50')
````

3. Machine Learning modelini o'qitish

Bu kod bizga haqiyqiy datasets uchun optimal chiziqli grafikni chizib berdi va title qismiga formulasini ham topib yozib qo'ydik.

````python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# data
data = make_regression(n_samples=40, n_features=1, noise=10, random_state=42)
x_data, y_data = data[0], data[1]

# Regressiyadan foydalanamiz
regression = LinearRegression().fit(x_data, y_data)
y_pred = regression.predict(x_data)


# grafikni chiqarish
plt.scatter(x_data, y_data, label='Haqiqiy data sets')
plt.plot(x_data, y_pred, 'r', label='ML dastur')
plt.xlabel('x')
plt.ylabel('Y')
plt.legend()
title = f"{regression.coef_[0]:3f}x {regression.intercept_:3f}"
plt.title(title)
````

4. Deep learning

Bu modelda oldingi modellarga nisbatan boshqacha natija chiqdi, ya'ni huddi bo'yab tashlan

````python
from sklearn.datasets import make_regression
from sklearn.neural_network import MLPRegressor
import matplotlib.pyplot as plt

# data
data = make_regression(n_samples=40, n_features=1, noise=10, random_state=42)
x_data, y_data = data[0], data[1]

# Regressiyadan foydalanamiz
regression = MLPRegressor(max_iter=2500).fit(x_data, y_data)
y_pred = regression.predict(x_data)


# grafikni chiqarish
plt.scatter(x_data, y_data, label='Haqiqiy data sets')
plt.plot(x_data, y_pred, 'r', label='DL dastur')
plt.xlabel('x')
plt.ylabel('Y')
plt.legend()
````
